La distorsión total de armónicos (THD, por sus siglas en inglés) es una medida de la cantidad de distorsión armónica presente en una señal eléctrica. Los armónicos son señales de frecuencias que son múltiplos enteros de la frecuencia fundamental de la señal. En un circuito eléctrico ideal, la señal debe ser pura (sin distorsión), pero en la práctica, la distorsión puede ocurrir debido a factores como componentes no lineales o cargas que generan corrientes no sinusoidales.

THD se calcula comparando la amplitud de los armónicos presentes en la señal con la amplitud de la frecuencia fundamental (la frecuencia principal). Se expresa generalmente como un porcentaje, y se calcula de la siguiente manera:
$$\text{THD} = \frac{\sqrt{V_2^2 + V_3^2 + \cdots + V_n^2}}{V_1} \times 100$$

donde:

- V1 es la amplitud de la frecuencia fundamental,
- V2,V3,…Vn son las amplitudes de los armónicos de orden 2, 3, hasta n

Un THD más bajo significa que la señal está más cerca de ser una onda pura, mientras que un THD alto indica que la señal está significativamente distorsionada.


Por otro lado, el factor de potencia (FP) es una medida de cuán eficientemente se está utilizando la energía en un circuito eléctrico. Es la relación entre la potencia activa (real) y la potencia aparente. En otras palabras, el factor de potencia indica cuánto de la energía que se suministra al sistema se convierte en trabajo útil.

Se calcula con la fórmula:
$$\text{FP} = \frac{P}{S}$$


donde:

- P es la potencia activa (real) en vatios (W),
- S es la potencia aparente en voltamperios (VA).

El factor de potencia puede estar entre 0 y 1. Un factor de potencia de 1 (o 100%) significa que toda la potencia suministrada al sistema se convierte en trabajo útil, lo cual es ideal. Un factor de potencia bajo indica que parte de la energía suministrada no se utiliza eficientemente, generalmente debido a la presencia de cargas inductivas o capacitivas en el sistema.

Para calcular la Distorsión Total de Armónicos  (THD) desde la Transformada Rápida de Fourier (FFT) y la distorsión del factor de potencia con base en el THD, se siguen algunos pasos y conceptos clave.

1. Cálculo del THD desde la FFT
La FFT descompone una señal en sus componentes de frecuencia, permitiéndote obtener las amplitudes de la frecuencia fundamental y sus armónicos. El THD se calcula utilizando la relación entre la amplitud de la frecuencia fundamental y la suma de las amplitudes de los armónicos.

Pasos:
1. Realiza la FFT de la señal:

- La FFT transforma la señal en el dominio de frecuencia, dándote los componentes espectrales (frecuencia fundamental y armónicos).
- El resultado de la FFT es una serie de amplitudes para cada frecuencia.

2. Identifica la amplitud de la frecuencia fundamental:

- La frecuencia fundamental es la primera componente en la FFT (por ejemplo, para una señal de corriente alterna de 50 Hz, la frecuencia fundamental es 50 Hz).

3. Identifica las amplitudes de los armónicos:

- Los armónicos son las componentes de frecuencia que son múltiplos enteros de la frecuencia fundamental (2 veces la frecuencia fundamental, 3 veces la frecuencia fundamental, etc.).
- Así, el armónico de orden 2 está en la frecuencia 100 Hz (en el caso de 50 Hz como fundamental), el de orden 3 en 150 Hz, etc.

4. Calcula el THD: Utiliza la fórmula del THD:

$$\text{THD} = \frac{\sqrt{V_2^2 + V_3^2 + \cdots + V_n^2}}{V_1} $$

donde:
- V1 es la amplitud de la frecuencia fundamental,
- V2,V3,…,Vn son las amplitudes de los armónicos de orden 2, 3, hasta n.

Ejemplo: Si la amplitud de la frecuencia fundamental 𝑉1=10V, la amplitud del armónico de orden 2 V2=2V y la del armónico de orden 3V3=1V, entonces:

$$\text{THD} = \frac{\sqrt{2^2 + 1^2}}{10}$$

$$\text{THD} = \frac{\sqrt{4 + 1}}{10}$$

$$\text{THD} = \frac{\sqrt{5}}{10} \approx 0.224 \, (22.4\%)$$


2. Distorsión del Factor de Potencia con Base en el THD
El factor de potencia (FP) se ve afectado por los armónicos presentes en la señal. Un THD alto indica una mayor distorsión de la señal y, por ende, puede reducir el factor de potencia, ya que parte de la energía se disipa en frecuencias no útiles.

La relación entre el THD y el factor de potencia no es directa, pero se puede inferir que a medida que el THD aumenta, el factor de potencia tiende a disminuir debido a la presencia de armónicos que generan corrientes reactivas (inductivas o capacitivas) que no contribuyen al trabajo útil.

Para calcular una aproximación de la distorsión en el factor de potencia (FP distorsionado) debido al THD, una forma común de relacionarlo es utilizando una corrección de factor de potencia que se ve influenciada por los armónicos.

Fórmula aproximada:
$$\text{FP}_{\text{distorsionado}} = \frac{\text{FP}_{\text{ideal}}}{\sqrt{1 + \text{THD}^2}}$$

Donde:

- FP ideal = es el factor de potencia sin distorsión (idealmente 1),

- THD es el valor de la distorsión total de armónicos (en forma decimal).

Ejemplo: Si el factor de potencia ideal es
1 y el THD es 22.4% = 0.224, entonces:

$$\text{FP}_{\text{distorsionado}} = \frac{1}{\sqrt{1 + (0.224)^2}}$$

$$\text{FP}_{\text{distorsionado}} = \frac{1}{\sqrt{1 + 0.0502}}$$

$$\text{FP}_{\text{distorsionado}} = \frac{1}{\sqrt{1.0502}}$$

$$\text{FP}_{\text{distorsionado}} \approx 0.975$$

- THD se calcula a partir de las amplitudes obtenidas en la FFT, comparando la amplitud de los armónicos con la frecuencia fundamental.
- La distorsión en el factor de potencia puede aproximarse considerando cómo el THD afecta la potencia activa, usando una fórmula que ajusta el factor de potencia según el valor de THD.

Estos cálculos ayudan a entender el impacto de los armónicos en la eficiencia energética de los sistemas eléctricos.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft

# Funciones auxiliares
def generar_senal_rectificada(frecuencia, resistencia, capacitancia, tiempo_total, fs):
    """
    Genera una señal de un rectificador de onda completa con carga RC o resistiva.
    """
    t = np.linspace(0, tiempo_total, int(fs * tiempo_total))
    senal_fuente = np.sin(2 * np.pi * frecuencia * t)

    # Simular carga resistiva o RC
    if capacitancia is None:  # Carga resistiva
        corriente = np.abs(senal_fuente) / resistencia
    else:  # Carga RC
        tau = resistencia * capacitancia
        corriente = np.zeros_like(t)
        for i in range(1, len(t)):
            corriente[i] = corriente[i - 1] + (senal_fuente[i] - corriente[i - 1]) / (tau * fs)
        corriente = np.abs(corriente)

    return t, senal_fuente, corriente

def calcular_fft(senal, fs):
    """
    Calcula la FFT y retorna las frecuencias y sus magnitudes.
    """
    N = len(senal)
    fft_result = fft(senal)
    freqs = np.fft.fftfreq(N, 1 / fs)[:N // 2]
    magnitudes = 2 / N * np.abs(fft_result[:N // 2])
    return freqs, magnitudes

def calcular_thd(magnitudes):
    """
    Calcula la THD a partir de las magnitudes de la FFT.
    """
    V1 = magnitudes[0] if len(magnitudes) > 0 else 0
    Vn = magnitudes[1:] if len(magnitudes) > 1 else np.array([0])
    THD = np.sqrt(np.sum(Vn**2)) / V1 if V1 > 0 else 0
    return THD * 100

# Parámetros de simulación
frecuencia = 60  # Frecuencia de la fuente (Hz)
resistencia = 150  # Resistencia (ohmios)
capacitancia = 500e-6  # Capacitancia (faradios), usar None para carga resistiva
tiempo_total = 1  # Duración de la simulación (s)
fs = 10000  # Frecuencia de muestreo (Hz)

# Generar señal
carga_tipo = "RC" if capacitancia else "Resistiva"
t, senal_fuente, corriente = generar_senal_rectificada(frecuencia, resistencia, capacitancia, tiempo_total, fs)

# Calcular FFT
freqs, magnitudes = calcular_fft(corriente, fs)

# Calcular THD
thd = calcular_thd(magnitudes)

# Resultados
print(f"Tipo de carga: {carga_tipo}")
print(f"Resistencia: {resistencia} ohmios")
print(f"Capacitancia: {capacitancia if capacitancia else 'N/A'} faradios")
print(f"THD: {thd:.2f}%")

# Gráficos
plt.figure(figsize=(12, 6))

# Señal de corriente
plt.subplot(2, 1, 1)
plt.plot(t, corriente, label="Corriente (Carga RC)" if capacitancia else "Corriente (Carga Resistiva)")
plt.title("Señal de corriente")
plt.xlabel("Tiempo (s)")
plt.ylabel("Corriente (A)")
plt.grid()
plt.legend()

# FFT
plt.subplot(2, 1, 2)
# Remove the use_line_collection argument
plt.stem(freqs, magnitudes, basefmt=" ")
plt.title("Espectro de frecuencias (FFT)")
plt.xlabel("Frecuencia (Hz)")
plt.ylabel("Magnitud")
plt.grid()
plt.tight_layout()
plt.show()
```
